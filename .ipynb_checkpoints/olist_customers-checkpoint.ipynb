```python
# -*- coding: utf-8 -*-
"""
Olist CRM 분석 - 최종 통합본
======================================================================
분석 여정 요약 (README 참고):
  1) 초기 가설: "배송 지연 -> VIP 이탈" (단순 비교로는 22.72% vs 21.51%로 보임)
  2) 코드 검증 중 두 가지 버그 발견:
       - item merge로 인한 order 중복 카운트
       - Frequency>1 을 재구매 지표로 써서 시간 순서를 무시한 문제
  3) 시간 순서(t0) + censoring을 반영해 재검증 -> 전체 비교에서는 효과가
     사라짐 (p=0.52)
  4) 지연 심각도(Critical 7일+/Warning 1~6일)로 세그먼트를 나눠 재검증
     -> 오히려 "경미한 지연(Warning)"에서 재구매율 하락이 일관되게 관찰됨
  5) 60/90/120/180일 window sensitivity + 부트스트랩 신뢰구간으로 견고성 확인
     -> Warning 효과는 방향이 4개 window 모두 동일 (일관된 신호)
        Critical 효과는 방향 자체가 window마다 바뀜 (노이즈에 가까움)

결론: "심각한 지연이 VIP를 이탈시킨다"는 원 가설은 기각.
      "경미한 지연이 오히려 방치되어 위험하다"는 재설계된 CRM 전략으로 이어짐.
======================================================================
"""

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from google.colab import drive

plt.rcParams['axes.unicode_minus'] = False

drive.mount('/content/drive')
path = '/content/drive/MyDrive/project2/'

WINDOW_DAYS = 90        # 기본 관찰기간 (재구매 인정 기간)
VIP_RFM_THRESHOLD = 12  # RFM 상위 20% 컷오프
CRITICAL_DELAY_DAYS = 7  # Critical/Warning 구분 기준


# =====================================================================
# 1. 데이터 로드 및 order 단위 통합 (item merge로 인한 중복 방지)
# =====================================================================
def load_data():
    orders = pd.read_csv(path + "olist_orders_dataset.csv")
    items = pd.read_csv(path + "olist_order_items_dataset.csv")
    payments = pd.read_csv(path + "olist_order_payments_dataset.csv")
    customers = pd.read_csv(path + "olist_customers_dataset.csv")
    reviews = pd.read_csv(path + "olist_order_reviews_dataset.csv")
    products = pd.read_csv(path + "olist_products_dataset.csv")
    category_trans = pd.read_csv(path + "product_category_name_translation.csv")
    return orders, items, payments, customers, reviews, products, category_trans


orders, items, payments, customers, reviews, products, category_trans = load_data()

pay_agg = payments.groupby("order_id")["payment_value"].sum().reset_index()
rev_agg = reviews.groupby("order_id")["review_score"].mean().reset_index()

# 주문의 대표 카테고리: 결제금액이 가장 큰 상품 라인의 카테고리로 정의 (order당 1개로 고정)
items_cat = items.merge(products[['product_id', 'product_category_name']], on='product_id', how='left')
items_cat = items_cat.merge(category_trans, on='product_category_name', how='left')
order_category = (
    items_cat.sort_values('price', ascending=False)
    .drop_duplicates(subset='order_id', keep='first')[['order_id', 'product_category_name_english']]
)

df = orders.merge(customers, on="customer_id", how="left")
df = df.merge(pay_agg, on="order_id", how="left")
df = df.merge(rev_agg, on="order_id", how="left")
df = df.merge(order_category, on="order_id", how="left")

assert df['order_id'].is_unique, "order 단위 dedup 실패 - merge 과정 재확인 필요"

date_cols = ['order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col])

df_del = df[df['order_status'] == 'delivered'].copy()
df_del['delay_days'] = (df_del['order_delivered_customer_date'] - df_del['order_estimated_delivery_date']).dt.days
df_del['is_late'] = df_del['delay_days'] > 0

print("=== [1] 배송 지연이 리뷰 점수에 미치는 영향 (order-level) ===")
print(df_del.groupby('is_late')['review_score'].mean())
print(f"Total Late Orders (order 단위, 중복 없음): {df_del['is_late'].sum()}")

plt.figure(figsize=(10, 5))
delay_summary = df_del[(df_del['delay_days'] >= -10) & (df_del['delay_days'] <= 20)]
delay_summary.groupby('delay_days')['review_score'].mean().plot(marker='o', color='red')
plt.axvline(x=0, color='black', linestyle='--')
plt.title('Review Score by Delivery Delay Days')
plt.xlabel('Delay Days')
plt.ylabel('Avg Review Score')
plt.grid(True)
plt.show()


# =====================================================================
# 2. RFM 세그멘테이션 (VIP 정의)
# =====================================================================
snapshot_date = df_del['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

rfm = df_del.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days,
    'order_id': 'nunique',
    'payment_value': 'sum'
})
rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm['R'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['F'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm['M'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])
rfm['RFM_Score'] = rfm[['R', 'F', 'M']].astype(int).sum(axis=1)

df_final = df_del.merge(rfm[['Frequency', 'RFM_Score']], on='customer_unique_id', how='left')
vip_mask = df_final['RFM_Score'] >= VIP_RFM_THRESHOLD

vip_orders = df_final[vip_mask].copy().sort_values(['customer_unique_id', 'order_purchase_timestamp'])
vip_orders['risk_level'] = np.where(
    vip_orders['delay_days'] >= CRITICAL_DELAY_DAYS, 'critical',
    np.where(vip_orders['is_late'], 'warning', 'none')
)


# =====================================================================
# 3. 시간 순서를 반영한 cohort 구성 함수 (재사용)
# =====================================================================
def build_cohort(vip_orders, snapshot_date, window_days):
    """세그먼트(critical/warning/on_time)별로 t0를 잡고, t0+window_days 이내
    재구매 여부를 계산. censoring(관찰기간 미충족 고객)은 제외."""
    first_critical = (
        vip_orders[vip_orders['risk_level'] == 'critical']
        .groupby('customer_unique_id')['order_purchase_timestamp'].min()
    )
    first_warning = (
        vip_orders[vip_orders['risk_level'] == 'warning']
        .groupby('customer_unique_id')['order_purchase_timestamp'].min()
    )
    first_order = vip_orders.groupby('customer_unique_id')['order_purchase_timestamp'].min()

    cohort = pd.DataFrame({'first_order': first_order})
    cohort = cohort.join(first_critical.rename('first_critical'))
    cohort = cohort.join(first_warning.rename('first_warning'))

    cohort['segment'] = np.select(
        [cohort['first_critical'].notna(), cohort['first_warning'].notna()],
        ['critical', 'warning'],
        default='on_time'
    )
    cohort['t0'] = cohort['first_critical']
    cohort['t0'] = cohort['t0'].fillna(cohort['first_warning'])
    cohort['t0'] = cohort['t0'].fillna(cohort['first_order'])

    cohort = cohort[cohort['t0'] + pd.Timedelta(days=window_days) <= snapshot_date].copy()

    orders_after = vip_orders.merge(cohort[['t0']], left_on='customer_unique_id', right_index=True)
    orders_after = orders_after[
        (orders_after['order_purchase_timestamp'] > orders_after['t0']) &
        (orders_after['order_purchase_timestamp'] <= orders_after['t0'] + pd.Timedelta(days=window_days))
    ]
    repurchase_customers = set(orders_after['customer_unique_id'].unique())
    cohort['is_retained_after_t0'] = cohort.index.isin(repurchase_customers)
    return cohort


def pairwise_chi2(cohort, group_a, group_b):
    sub = cohort[cohort['segment'].isin([group_a, group_b])]
    table = pd.crosstab(sub['segment'], sub['is_retained_after_t0'])
    if table.shape[0] < 2 or table.shape[1] < 2:
        return np.nan, np.nan, table
    chi2, p, dof, expected = chi2_contingency(table)
    return chi2, p, table


# =====================================================================
# 4. 세그먼트별 재구매율 비교 (기본 window=90일)
# =====================================================================
cohort_90 = build_cohort(vip_orders, snapshot_date, WINDOW_DAYS)
seg_counts = cohort_90['segment'].value_counts()
retention_by_seg = cohort_90.groupby('segment')['is_retained_after_t0'].mean() * 100

print(f"\n=== [2] 세그먼트별 시간 순서 기반 재구매율 (window={WINDOW_DAYS}일, n={len(cohort_90)}) ===")
print(seg_counts)
print(retention_by_seg)

for other in ['warning', 'critical']:
    chi2, p, table = pairwise_chi2(cohort_90, 'on_time', other)
    print(f"\n[on_time vs {other}] chi2={chi2:.3f}, p-value={p:.4f}")
    print(table)

order_cols = ['on_time', 'warning', 'critical']
retention_plot = retention_by_seg.reindex(order_cols)
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=retention_plot.index, y=retention_plot.values,
                  hue=retention_plot.index, palette='Set2', legend=False)
plt.title(f'Retention Rate by Delay Severity (t0 + {WINDOW_DAYS}d window)')
plt.ylabel('Repurchase Rate (%)')
plt.ylim(0, max(retention_plot.values) + 5)
for p_ in ax.patches:
    ax.annotate(f'{p_.get_height():.2f}%', (p_.get_x() + p_.get_width() / 2., p_.get_height()),
                ha='center', va='center', xytext=(0, 9), textcoords='offset points', fontweight='bold')
plt.show()


# =====================================================================
# 5. Window Sensitivity 분석 (60/90/120/180일)
# =====================================================================
windows_to_test = [60, 90, 120, 180]
sensitivity_rows = []
for w in windows_to_test:
    c = build_cohort(vip_orders, snapshot_date, w)
    sc = c['segment'].value_counts()
    r = c.groupby('segment')['is_retained_after_t0'].mean() * 100
    _, p_warn, _ = pairwise_chi2(c, 'on_time', 'warning')
    _, p_crit, _ = pairwise_chi2(c, 'on_time', 'critical')
    sensitivity_rows.append({
        'window_days': w, 'n_total': len(c),
        'n_on_time': sc.get('on_time', 0), 'n_warning': sc.get('warning', 0), 'n_critical': sc.get('critical', 0),
        'retention_on_time(%)': round(r.get('on_time', np.nan), 2),
        'retention_warning(%)': round(r.get('warning', np.nan), 2),
        'retention_critical(%)': round(r.get('critical', np.nan), 2),
        'p_warning_vs_ontime': round(p_warn, 4) if not np.isnan(p_warn) else np.nan,
        'p_critical_vs_ontime': round(p_crit, 4) if not np.isnan(p_crit) else np.nan,
    })

sensitivity_df = pd.DataFrame(sensitivity_rows)
print("\n=== [3] Window Sensitivity 분석 ===")
print(sensitivity_df.to_string(index=False))

plt.figure(figsize=(9, 5))
plt.plot(sensitivity_df['window_days'], sensitivity_df['retention_on_time(%)'], marker='o', label='on_time')
plt.plot(sensitivity_df['window_days'], sensitivity_df['retention_warning(%)'], marker='o', label='warning')
plt.plot(sensitivity_df['window_days'], sensitivity_df['retention_critical(%)'], marker='o', label='critical')
plt.xlabel('Observation Window (days)')
plt.ylabel('Repurchase Rate (%)')
plt.title('Retention Rate Sensitivity by Observation Window')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

plt.figure(figsize=(9, 4))
plt.plot(sensitivity_df['window_days'], sensitivity_df['p_warning_vs_ontime'], marker='s',
          color='darkorange', label='warning vs on_time')
plt.plot(sensitivity_df['window_days'], sensitivity_df['p_critical_vs_ontime'], marker='s',
          color='steelblue', label='critical vs on_time')
plt.axhline(y=0.05, color='red', linestyle='--', label='p = 0.05 threshold')
plt.xlabel('Observation Window (days)')
plt.ylabel('p-value')
plt.title('Statistical Significance by Observation Window')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()


# =====================================================================
# 6. Warning 세그먼트 부트스트랩 신뢰구간
# =====================================================================
np.random.seed(42)
N_BOOTSTRAP = 2000

warning_vals = cohort_90.loc[cohort_90['segment'] == 'warning', 'is_retained_after_t0'].values
ontime_vals = cohort_90.loc[cohort_90['segment'] == 'on_time', 'is_retained_after_t0'].values


def bootstrap_mean_ci(values, n_boot=2000, ci=95):
    means = [np.random.choice(values, size=len(values), replace=True).mean() for _ in range(n_boot)]
    lower = np.percentile(means, (100 - ci) / 2)
    upper = np.percentile(means, 100 - (100 - ci) / 2)
    return np.mean(values), lower, upper, means


warn_mean, warn_lo, warn_hi, warn_dist = bootstrap_mean_ci(warning_vals, N_BOOTSTRAP)
on_mean, on_lo, on_hi, on_dist = bootstrap_mean_ci(ontime_vals, N_BOOTSTRAP)

print(f"\n=== [4] 부트스트랩 95% 신뢰구간 (window={WINDOW_DAYS}일, n_boot={N_BOOTSTRAP}) ===")
print(f"warning : {warn_mean*100:.2f}%  (95% CI: {warn_lo*100:.2f}% ~ {warn_hi*100:.2f}%)  n={len(warning_vals)}")
print(f"on_time : {on_mean*100:.2f}%  (95% CI: {on_lo*100:.2f}% ~ {on_hi*100:.2f}%)  n={len(ontime_vals)}")

plt.figure(figsize=(7, 5))
plt.hist(np.array(warn_dist) * 100, bins=40, alpha=0.6, label='warning (bootstrap means)')
plt.hist(np.array(on_dist) * 100, bins=40, alpha=0.6, label='on_time (bootstrap means)')
plt.axvline(warn_mean * 100, color='tab:blue', linestyle='--')
plt.axvline(on_mean * 100, color='tab:orange', linestyle='--')
plt.xlabel('Repurchase Rate (%)')
plt.ylabel('Bootstrap Frequency')
plt.title('Bootstrap Distribution: Warning vs On-time Retention Rate')
plt.legend()
plt.show()


# =====================================================================
# 7. 카테고리별 위험도 (order-level)
# =====================================================================
vip_late = df_final[vip_mask & df_final['is_late']].copy()
vip_late['risk_level'] = vip_late['delay_days'].apply(lambda x: 'Critical' if x >= CRITICAL_DELAY_DAYS else 'Warning')

print(f"\n=== [5] 카테고리별 위험도 (지연 겪은 VIP 주문 수: {len(vip_late)}) ===")
top_10_late = vip_late['product_category_name_english'].value_counts().head(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_late.values, y=top_10_late.index, hue=top_10_late.index, palette='viridis', legend=False)
plt.title('Top 10 Critical Categories for VIPs (order-level)')
plt.show()


def generate_crm_msg(row):
    cat_name = row['product_category_name_english'] if pd.notnull(row['product_category_name_english']) else 'General'
    if row['risk_level'] == 'Critical':
        return f"[{cat_name}] 지연 사과: VIP 20% 특별 할인권 발송"
    return f"[{cat_name}] 지연 안내: 감사 10% 쿠폰 발송"


vip_late['personalized_message'] = vip_late.apply(generate_crm_msg, axis=1)
print("\n--- [기존] 개인화 오퍼 샘플 (Critical 중심 설계 - 아래 8절에서 재검토) ---")
print(vip_late[['customer_unique_id', 'risk_level', 'personalized_message']].head())


# =====================================================================
# 8. [재설계] Recovery Campaign ROI 시뮬레이션
#
#    기존 가정: Critical(7일+)에 더 큰 보상(20%), Warning(1~6일)에 작은 보상(10%)
#    -> 4~7절의 검증 결과, 이 배분은 데이터와 반대 방향임이 드러남:
#       Warning 세그먼트에서만 일관된 재구매율 하락이 관찰되고,
#       Critical 세그먼트는 window마다 방향조차 바뀌는 노이즈에 가까웠음.
#
#    따라서 이 섹션은 "Critical 우선"이 아니라 "Warning 우선" 예산 배분을
#    검토하기 위한 시뮬레이션으로 재설계함. 전환율 가정치는 기존과 동일하게
#    보수적 추정치이며, 실제 캠페인 집행 후 A/B 테스트로 검증이 필요함.
# =====================================================================
warning_orders = vip_late[vip_late['risk_level'] == 'Warning']
critical_orders = vip_late[vip_late['risk_level'] == 'Critical']

print(f"\n=== [6] 재설계된 ROI 시뮬레이션 (세그먼트별 예산 분리) ===")
print(f"Warning 주문 수: {len(warning_orders)}, 평균 결제액: {warning_orders['payment_value'].mean():.2f} BRL")
print(f"Critical 주문 수: {len(critical_orders)}, 평균 결제액: {critical_orders['payment_value'].mean():.2f} BRL")


def simulate_roi(segment_df, coupon, conv_rate, label):
    n = len(segment_df)
    avg_value = segment_df['payment_value'].mean()
    cost = n * coupon
    revenue = (n * conv_rate) * avg_value
    return {
        'Segment': label, 'N_Orders': n, 'Coupon(BRL)': coupon, 'Conv_Rate(%)': conv_rate * 100,
        'Total_Cost': round(cost, 2), 'Defended_Revenue': round(revenue, 2),
        'ROI': round(revenue / cost, 2) if cost > 0 else np.nan
    }


# 재설계안: 데이터 근거가 있는 Warning에 더 높은 보상(15 BRL), 근거가 약한 Critical은
# 기존보다 보수적인 보상(10 BRL)으로 재배분한 시나리오 예시
revised_scenarios = [
    simulate_roi(warning_orders, coupon=15, conv_rate=0.20, label='Warning (재설계: 우선순위 상향)'),
    simulate_roi(critical_orders, coupon=10, conv_rate=0.15, label='Critical (재설계: 우선순위 하향)'),
]
revised_roi_df = pd.DataFrame(revised_scenarios)
print(revised_roi_df.to_string(index=False))

print("""
NOTE:
- 전환율(conv_rate) 가정치는 실제 과거 쿠폰 캠페인 데이터가 없어 보수적으로 가정한 값.
  포트폴리오/README에는 반드시 "가정치"임을 명시할 것.
- 이 재배분안은 4~7절의 발견(Warning=일관된 신호, Critical=노이즈에 가까움)에 근거한
  제안이며, 실제 적용 전 소규모 A/B 테스트로 검증이 필요함.
""")
```
